```python
def fill_missing_values(df):
    # 数値型の欠損値は平均値で埋める
    num_cols = df.select_dtypes(include=['number']).columns
    df[num_cols] = df[num_cols].fillna(round(df[num_cols].mean()))
    
    # カテゴリ型の欠損値は最頻値で埋める
    cat_cols = df.select_dtypes(include=['category']).columns
    df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode())
    
    # オブジェクト型（文字列型）の欠損値は 'Unknown' で埋める
    obj_cols = df.select_dtypes(include=['object']).columns
    df[obj_cols] = df[obj_cols].fillna('Unknown')
    
    return df
```

```python
df_preprocessed = fill_missing_values(df_preprocessed) 
```

```python
df_preprocessed_train = df_preprocessed.iloc[:df_train.shape[0], :]
df_preprocessed_test = df_preprocessed.iloc[df_train.shape[0]:, :]
```

# 型、項目数、欠損値率

```python
dtypes = []
val_counts_train = []
val_counts_test = []
isnull_train = []
isnull_test = []
isnull_train_ratio = 100 * df_train.isnull().sum() / len(df_train)
isnull_test_ratio = 100 * df_test.isnull().sum() / len(df_test)
for col in df_test.columns:
    dtypes.append(str(df_train[col].dtype))
    val_counts_train.append(len(df_train[col].value_counts()))
    val_counts_test.append(len(df_test[col].value_counts()))
    isnull_train.append(isnull_train_ratio[col])
    isnull_test.append(isnull_test_ratio[col])
```

```python
inds = ["型", "val_counts_train", "val_counts_test", "NaN率_train", "NaN率_test"]
df_eda = pd.DataFrame([dtypes, val_counts_train, val_counts_test, isnull_train, isnull_test], columns=df_test.columns, index=inds).T
df_eda.query("val_counts_train > 1")
```
